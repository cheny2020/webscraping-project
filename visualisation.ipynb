```python
import re
import json
import matplotlib.pyplot as plt
import collections
#pip install tabulate
from tabulate import tabulate
import pandas as pd
```

```python
data_europea = []
with open("europea.json","r") as file:
    data_europea = json.load(file)
```

```python
len(data_europea)
```

```python
data_bnf = []
with open("bnf.json","r") as file:
    data_bnf = json.load(file)
```

Europeana

['provider'] : pie chart

['title'] : show in table

['type'] : pie chart

['year'] : bar chart

```python
#europeana, get all provider
provider_dict={}
for i in range(0,len(data_europea)):
    if 'provider' in data_europea[i]:
        key=data_europea[i]['provider'][0]
        if key in provider_dict:
            provider_dict[key]+=1
        else:
            provider_dict[key]=1
labels = []
sizes = []

for x, y in provider_dict.items():
    labels.append(x)
    sizes.append(y)
    
plt.pie(sizes, labels=labels,autopct='%1.1f%%')

plt.axis('equal')
plt.title('Europeana results by provider')
plt.show()
```

```python
title_dict={}
for i in range(0,len(data_europea)):
    if 'title' in data_europea[i]:
        title=data_europea[i]['title'][0]
        if title in title_dict:
            title_dict[title]+=1
        else:
            title_dict[title]=1
#dict(sorted(title_dict.items(), key=lambda item: item[1],reverse=True))
```

```python
#europeana, get all type
type_dict={}
for i in range(0,len(data_europea)):
    if 'type' in data_europea[i]:
        key=data_europea[i]['type']
        if key in type_dict:
            type_dict[key]+=1
        else:
            type_dict[key]=1

labels = []
sizes = []

for x, y in type_dict.items():
    labels.append(x)
    sizes.append(y)
    
plt.pie(sizes, labels=labels,autopct='%1.1f%%')

plt.axis('equal')
plt.title('Europeana results by type')
plt.show()
```

```python
#europeana, get all year
year_dict={}
for i in range(0,len(data_europea)):
    if 'year' in data_europea[i]:
        key=data_europea[i]['year'][0]
        if key in year_dict:
            year_dict[key]+=1
        else:
            year_dict[key]=1

sorted_year_dict=collections.OrderedDict(sorted(year_dict.items()))

plt.bar(range(len(sorted_year_dict)), list(sorted_year_dict.values()), align='center')
plt.xticks(range(len(sorted_year_dict)), list(sorted_year_dict.keys()))
plt.ylabel('values')
plt.title('Europeana results by year')
plt.show()
```

```python
#europeana, show result in table
data=[]
for i in range(0,len(data_europea)):
    element=[]
    element.append(i+1)
    
    if 'title' in data_europea[i]:
        title=data_europea[i]['title'][0]
        title=title[0:40]
        if len(title)==40:
            title+="..."
        element.append(title)
    else:
        element.append(" ")

    if 'provider' in data_europea[i]:
        provider=data_europea[i]['provider'][0]
        element.append(provider)
    else:
        element.append(" ")        
        
    if 'type' in data_europea[i]:
        type=data_europea[i]['type']
        element.append(type)
    else:
        element.append(" ")    
        
    if 'year' in data_europea[i]:
        year=data_europea[i]['year'][0]
        element.append(year)
    else:
        element.append(" ")    

    if 'guid' in data_europea[i]:
        year=data_europea[i]['guid']
        element.append(year)
    else:
        element.append(" ")
        
    data.append(element)
print(tabulate(data, headers=["title", "provider", "type", "year","link"]))
```

```python
year_dict2={}
for i in range(0,len(data_bnf)):
    if 'date' in data_bnf[i]:
        date=data_bnf[i]['date'][1]
        #print(date)
        year=re.match(r'.*([1-3][0-9]{3})', date)
        if year is not None:
            #print(year.group(1))
            key=year.group(1)
            if key in year_dict2:
                year_dict2[key]+=1
            else:
                year_dict2[key]=1
#print(year_dict2)

sorted_year_dict2=collections.OrderedDict(sorted(year_dict2.items()))

plt.bar(range(len(sorted_year_dict2)), list(sorted_year_dict2.values()), align='center')
plt.xticks(range(len(sorted_year_dict2)), list(sorted_year_dict2.keys()))
plt.ylabel('values')
plt.title('BNF results by year')
plt.show()
```

```python
#bnf, get all type
type_dict2={}
for i in range(0,len(data_bnf)):
    if 'typologie' in data_bnf[i]:
        key=data_bnf[i]['typologie'][1]
        if key in type_dict2:
            type_dict2[key]+=1
        else:
            type_dict2[key]=1
labels = []
sizes = []

for x, y in type_dict2.items():
    labels.append(x)
    sizes.append(y)
    
plt.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.axis('equal')
plt.title('BNF results by type')
plt.show()
```

```python
#BNF, show result in table
data2=[]
for i in range(0,len(data_bnf)):
    element=[]
    element.append(i+1)
    
    if 'titre' in data_bnf[i]:
        title=data_bnf[i]['titre'][1][:40]
        element.append(title)
    else:
        element.append(" ")

    if 'auteur' in data_bnf[i]:
        auteur=data_bnf[i]['auteur'][1][:20]
        element.append(auteur)
    else:
        element.append(" ")        
        
    if 'publication' in data_bnf[i]:
        publication=data_bnf[i]['publication'][1][:20]
        element.append(publication)
    else:
        element.append(" ")    
        
    if 'editeur' in data_bnf[i]:
        editeur=data_bnf[i]['editeur'][1][:20]
        element.append(editeur)
    else:
        element.append(" ")        
        
    if 'date' in data_bnf[i]:
        date=data_bnf[i]['date'][1]
        year=re.match(r'.*([1-3][0-9]{3})', date)
        element.append(year.group(1))
    else:
        element.append(" ")    

    if 'typologie' in data_bnf[i]:
        type=data_bnf[i]['typologie'][1]
        element.append(type)
    else:
        element.append(" ")
        
    data2.append(element)
print(tabulate(data2, headers=["title", "author", "publication", "editor","year","type"]))
```
